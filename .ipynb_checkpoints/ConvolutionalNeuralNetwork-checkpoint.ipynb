```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
import os
all_dfs = []
directory = os.fsencode("face data/")
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith("0.csv"): 
        df = pd.read_csv("face data/" + filename, header=None)
        if (filename.endswith("10.csv")):
            df.loc[df[0] == 10, 0] = 1
        all_dfs.append(df)
len(all_dfs)
```

```python
data = pd.concat(all_dfs)
pd.set_option('display.max_rows', 10)
data
```

```python
X = data.iloc[:, [i for i in range(1, 3137)]]
Y = data.iloc[:, 0]
X
```

```python
X_Data = []
for _, row in X.iterrows():
    X_Data.append(np.array(row).reshape(56, 56))
```

```python
for i in range(len(X_Data)):
    plt.imshow(X_Data[i], cmap="gray")
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Data, Y, test_size = 0.20, shuffle=False)
```

```python
X_Train, X_Val, Y_Train, Y_Val = train_test_split(X_Data, Y, train_size=0.9, test_size=0.1, shuffle=False)
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Train, Y_Train, train_size=0.78, shuffle=False)
```

```python
X_Train = np.array(X_Train).reshape(len(X_Train), 56, 56, 1)
X_Val = np.array(X_Val).reshape(len(X_Val), 56, 56, 1)
X_Test = np.array(X_Test).reshape(len(X_Test), 56, 56, 1)
Y_Train = np.array(Y_Train)
Y_Val = np.array(Y_Val)
Y_Tests = np.array(Y_Test)
X_Train.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (2,2), activation='relu', input_shape=(56, 56, 1)),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Conv2D(64, (2,2), activation='relu', kernel_regularizer=tf.keras.regularizers.l2(l=0.01)),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(250, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_Train, Y_Train, epochs=10, validation_data=(X_Val, Y_Val))
```

```python
model.evaluate(X_Test, Y_Test)
```

```python
Y_Pred = np.argmax(model.predict(X_Test), axis=-1)
both = np.vstack((np.array(Y_Test), Y_Pred)).T.astype(np.int)
both
```

```python
con_mat = tf.math.confusion_matrix(labels=Y_Test, predictions=Y_Pred).numpy()
con_mat_norm = np.around(con_mat.astype('float') / con_mat.sum(axis=1)[:, np.newaxis], decimals=2)
con_mat_norm
```
