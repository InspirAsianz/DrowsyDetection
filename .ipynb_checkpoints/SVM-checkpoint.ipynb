```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
import os
all_dfs = []
directory = os.fsencode("feature data/")
for file in os.listdir(directory):
     filename = os.fsdecode(file)
     if filename.endswith("0.csv"): 
         all_dfs.append(pd.read_csv("feature data/" + filename, header=None))
```

```python
df1_0 = pd.read_csv("data/01-0.csv", header=None)
df1_5 = pd.read_csv("data/01-5.csv", header=None)
df1_10 = pd.read_csv("data/01-10.csv", header=None)
df2_0 = pd.read_csv("data/02-0.csv", header=None)
df2_5 = pd.read_csv("data/02-5.csv", header=None)
df2_10 = pd.read_csv("data/02-10.csv", header=None)
```

```python
data = pd.concat(all_dfs) # [df1_0, df1_5, df1_10, df2_0, df2_5, df2_10])
data
```

```python
X = data.iloc[:, [i for i in range(1, 7)]]
Y = data.iloc[:, 0]
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
X1 = scaler.fit_transform(np.array(X.loc[:, 1].copy()).reshape(-1, 1))
X.iloc[:, 0] = X1
X2 = scaler.fit_transform(np.array(X.loc[:, 2].copy()).reshape(-1, 1))
X.iloc[:, 0] = X2
X3 = scaler.fit_transform(np.array(X.loc[:, 3].copy()).reshape(-1, 1))
X.iloc[:, 0] = X3
X4 = scaler.fit_transform(np.array(X.loc[:, 4].copy()).reshape(-1, 1))
X.iloc[:, 0] = X4
X5 = scaler.fit_transform(np.array(X.loc[:, 5].copy()).reshape(-1, 1))
X.iloc[:, 0] = X5
X6 = scaler.fit_transform(np.array(X.loc[:, 6].copy()).reshape(-1, 1))
X.iloc[:, 0] = X6
X.columns = ["Left Width", "Left Height", "Right Width", "Right Height", "Mouth Width", "Mouth Height"]
X
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.20, shuffle=False)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X2 = pca.fit_transform(X)
X_Train2 = pca.fit_transform(X_Train)
X_Test2 = pca.fit_transform(X_Test)
```

```python
zeroes = []
fives = []
tens = []
for i in range(0, len(Y)):
    if (Y.iloc[i] == 0):
        zeroes.append([X2[i][0], X2[i][1]])
    if (Y.iloc[i] == 10):
        tens.append([X2[i][0], X2[i][1]])

zeroes = np.array(zeroes)
tens = np.array(tens)

plt.scatter(zeroes[:, 0], zeroes[:, 1], color="blue")
plt.scatter(tens[:, 0], tens[:, 1], color="green")
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "poly")
classifier.fit(X_Train, Y_Train)
```

```python
from sklearn.metrics import accuracy_score
Y_Pred = classifier.predict(X_Test)
accuracy = accuracy_score(Y_Test, Y_Pred)
accuracy
```

```python
pd.concat([Y, X], axis=1).to_csv("feature data/scaled.csv", index=False)
```
